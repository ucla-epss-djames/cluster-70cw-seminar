```python
# data analysis library
import pandas as pd
# plotting library
import matplotlib.pyplot as plt
```

```python
# reading in the data
data = pd.read_csv("data/temp_data.csv")

# this is just to show the container in the cell below
data
```

```python
# defining variables to work with what's in the container
x = data["time"]
y = data["temp"]

# plotting out the data
plt.plot(x, y)
```

```python
# adding labels and changing the colors of the plot
plt.plot(x, y, color="red")
plt.xlabel("time [s]")
plt.ylabel("temp [C]")
plt.title("Temperature Time Series")
plt.grid(True)
```
